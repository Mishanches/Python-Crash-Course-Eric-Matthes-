## Случайное блуждание


```python
from random import choice
class RandomWalk():
    """Класс для генерирования случайных блужданий"""
    def __init__(self, num_points=5000):
        """Инициализирует атрибуты блуждания"""
        self.num_points = num_points
        """все блужждания начинаются с точки (0,0)"""
        self.x_values = [0]
        self.y_values = [0]
    
    def fill_walk(self):
        """Вычисляет все точки блуждания"""
        # шаги генерируются до достижения нужной длины
        while len(self.x_values) < self.num_points:
            #определение направляения и длины перемещения
            x_direction = choice([1,-1])
            x_distance = choice([0,1,2,3,4])
            x_step = x_direction * x_distance
            
            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4])
            y_step = y_direction * y_distance
            
            #отклонение нулевых перемещений
            if x_step == 0 and y_step == 0:
                continue
                
            #вычисление следующих значений x и y
            next_x = self.x_values[-1] + x_step
            next_y = self.y_values[-1] + y_step
            
            self.x_values.append(next_x)
            self.y_values.append(next_y)


```

```python
import matplotlib.pyplot as plt
#from random_walk import RandomWalk
#построение случайного блуждания и нанесение точек на диаграмму
rw = RandomWalk()
rw.fill_walk()
plt.scatter(rw.x_values, rw.y_values, s=15)
plt.show()
```

```python
from random import randint
class Die():
    """Класс, представляющий одиин кубик"""
    def __init__(self, num_sides=6):
        """по умолчанию 6 граней"""
        self.num_sides=num_sides
        
    def roll(self):
        """возвращает случайное число от 1 до n граней"""
        return randint(1, self.num_sides)
```
